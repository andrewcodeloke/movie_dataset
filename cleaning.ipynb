```python
import pandas as pd 
import re
df = pd.read_csv('movies.csv')
df
```

<h1> DATA CLEANING

```python
# viewing first 5 rows
df.head(5)
```

```python
# viewing last 5 rows
df.tail(5)
```

```python
# checking data types
df.dtypes
```

```python
# checking null values
df.isna().sum()
```

```python
# dropping rows with null values
# all Series Shows have null values
# can drop nulls to extract only movie related data
df = df.dropna()
df
```

```python
# resetting index
df = df.reset_index(drop=True)
df
```

```python
# removing /n from columns
df['GENRE'] = df['GENRE'].str.replace('\n','')
df['STARS'] = df['STARS'].str.replace('\n','')
df['ONE-LINE'] = df['ONE-LINE'].str.replace('\n','')
df
```

```python
# splitting Stars column into Director & Stars column
df[['DIRECTOR', 'STARS']] = df['STARS'].str.split('|', n=1, expand=True)
df
```

```python
# Remove prefixes from columns
df['DIRECTOR'] = df['DIRECTOR'].str.replace(r'Director:', '').str.strip()
df['DIRECTOR'] = df['DIRECTOR'].str.replace(r'Directors:', '').str.strip() # r is regular expression for raw string
df['STARS'] = df['STARS'].str.replace(r'Stars:', '').str.strip() # str.strip just removes blanks spaces
df
```

```python
# renaming Run Time column to add units 
df = df.rename(columns={'RunTime':'RUN TIME (Minutes)'})
df = df.rename(columns={'Gross':'GROSS'})
df
```

```python
# changing RUN TIME column to an int data type
df['RUN TIME (Minutes)'] = df['RUN TIME (Minutes)'].astype('int')
df.dtypes
```

```python
# converting GROSS column into float data type
df['GROSS'] = df['GROSS'].str.replace('$','',regex=False)
df['GROSS'] = df['GROSS'].str.replace('M','',regex=False)
df['GROSS'] = df['GROSS'].astype('float')
```

```python
# converting VOTES column into int data type
df['VOTES'] = df['VOTES'].str.replace(',','',regex=False)
df['VOTES'] = df['VOTES'].astype('int')
```

```python
# converting YEAR column to int data type
df['YEAR'] = df['YEAR'].str.extract(r'(\d+)').astype('Int64')
```

```python
# converting to csv format
df.to_csv('cleaned_data.csv',index=False)
```
