```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

# importing cleaned dataset
df = pd.read_csv('cleaned_data.csv')
```

<H1> DATA ANALYSIS

<h2> Analysis Objectives
    
1. Top 5 Highest Grossing Films (Bar)
2. Trend of Movies Released by Year (Line)
3. Top 5 Highest Rated Films
4. Top 5 Most Published Directors
5. Most Occured Words in Movie Titles

<h4> Results and conclusions made from the analysis are based off the limited data in this movie list dataset, which is data dated from 1941-2020.

<h2> 1. Top 5 Highest Grossing Films

```python
highest_grossing = df.sort_values(by='GROSS',ascending=False).head(5)
highest_grossing
```

```python
colors = ['red','blue','green','orange','yellow']
plt.bar(highest_grossing['MOVIES'],highest_grossing['GROSS'],color=colors)
plt.xticks(rotation='vertical')
plt.xlabel('Movie Names')
plt.ylabel('Gross Earnings (Millions)')
plt.title('Top 5 Highest Grossing Films')
plt.show()
```

<h2> 2. Trend of Movies Released by Year

```python
myear= df['YEAR'].value_counts().reset_index() # reset so YEAR isn't the index
myear.columns = ['YEAR', 'COUNT']
myear
```

```python
# sorting years in ascending order
myears = myear.sort_values(by='YEAR',ascending=True)
```

```python
plt.figure(figsize=(30, 10))
plt.plot(myears['YEAR'], myears['COUNT'], marker='o',color='purple')
plt.xticks(size=12)
plt.title('Number of Movies Published by Year',size=16)
plt.xlabel('Year',size=16)
plt.ylabel('Number of Movies Published',size=16)
plt.show()
```

<h2> 3. Top 5 Highest Rated Films

```python
top_five = df.sort_values(by='RATING',ascending=False).head(5)
top_five
```

```python
plt.bar(top_five['MOVIES'],top_five['RATING'],color='green')
plt.xticks(rotation='vertical')
plt.title('Top 5 Rated Films')
plt.xlabel('Movie Name')
plt.ylabel('Rating')
plt.show()
```

<h2> 4. Top 5 Most Published Directors

```python
direc = df['DIRECTOR'].value_counts().head(5)
direc
```

```python
# turning results into a data frame
direc_pd = pd.DataFrame((direc).items(),columns=['DIRECTOR','OCCURENCES'])
direc_pd
```

```python
plt.bar(direc_pd['DIRECTOR'],direc_pd['OCCURENCES'],color='magenta')
plt.title('Top 5 Most Published Directors')
plt.xticks(rotation='vertical')
plt.xlabel('Director Name')
plt.ylabel('Occurences')
plt.show()
```

<h2> 5. Most Occurred Words in Movie Titles

```python
df
```

```python
# spltting each movie title based on space
split = df['MOVIES'].str.split(pat=' ')
split
```

```python
# empty list to store
flattened_list = [] 

# flattening the list using for loops
for title in split :
    for word in title :
        flattened_list.append(word)
```

```python
# finding most occured words in the Movie Titles
list_pd = pd.DataFrame(flattened_list)
lse = list_pd.value_counts().head(20)
lse
```
